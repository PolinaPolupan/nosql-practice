```python
import json
from jsonschema import Draft202012Validator
```

```python
schema = {
  "$schema": "https://json-schema.org/draft/2020-12/schema",
  "title": "E-Commerce Schema",
  "type": "object",
  "required": ["users", "products", "orders"],
  "properties": {
    "users": {
      "type": "array",
      "items": {
        "type": "object",
        "required": ["id", "name", "email", "registered_at"],
        "properties": {
          "id": { "type": "integer" },
          "name": { "type": "string" },
          "email": { "type": "string", "format": "email" },
          "registered_at": { "type": "string", "format": "date" }
        }
      }
    },

    "products": {
      "type": "array",
      "items": {
        "type": "object",
        "required": ["id", "name", "price", "tags"],
        "properties": {
          "id": { "type": "integer" },
          "name": { "type": "string" },
          "price": { "type": "number", "minimum": 0 },
          "tags": {
            "type": "array",
            "items": { "type": "string" }
          }
        }
      }
    },

    "orders": {
      "type": "array",
      "items": {
        "type": "object",
        "required": ["id", "user_id", "product_ids", "total"],
        "properties": {
          "id": { "type": "integer" },
          "user_id": { "type": "integer" },
          "product_ids": {
            "type": "array",
            "items": { "type": "integer" },
            "minItems": 1
          },
          "total": { "type": "number", "minimum": 0 }
        }
      }
    }
  }
}
```

```python
validator = Draft202012Validator(schema)
```

```python
with open("../data/2/valid.json") as f:
    valid_data = json.load(f)

with open("../data/2/invalid.json") as f:
    invalid_data = json.load(f)
```

```python
print("=== VALID DATA CHECK ===")
errors = list(validator.iter_errors(valid_data))

if not errors:
    print("Valid data passed")
else:
    for e in errors:
        print(e.message)
```

```python
print("=== INVALID DATA CHECK ===")
errors = list(validator.iter_errors(invalid_data))

if not errors:
    print("Invalid data unexpectedly passed")
else:
    for e in errors:
        print("Error:", e.message)
```
